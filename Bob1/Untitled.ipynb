```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
L = np.genfromtxt("L.txt", delimiter=",")
X1 = np.genfromtxt("X1.txt", delimiter=",")
FX1 = np.genfromtxt("FX1.txt", delimiter=",")
```

```python
pairs = [(X1[i],FX1[i], L[i]) for i in range(X1.shape[0])]
```

```python
def sort_key(pair):
    return pair[0]

pairs = sorted(pairs, key = sort_key)
```

```python
plt.figure()
plt.plot([pair[0] for pair in pairs],[pair[1] for pair in pairs])
plt.scatter([pair[0] for pair in pairs],[pair[2] for pair in pairs], color = 'r')
##plt.scatter(X1,L)
plt.show()
```
